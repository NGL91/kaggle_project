# Data Handling

```python
df = pd.read_csv('train.csv')
df.head(10)
```

### Explore data

```python
print (df.shape)
for name in df.columns.values:
    total_num = df[name].isnull().sum()
    if total_num > 0:
        print ('Number of null of value of {0}: {1}'.format(name,total_num))
```

As we see above, our training dataset have 12 columns and 891 rows.
There are some columns with a lot of NaN value consist of Age, Cabin.
The data cabin has more than a half of value is NaN, so maybe we will drop col Cabin. add later drop all row have NaN value.



```python
#Remove col Cabin
df.drop(['Cabin'],axis=1, inplace=True)
#Remove all rows have NaN value
df.dropna(inplace=True)
```

```python
#Check again
print (df.shape)
for name in df.columns.values:
    total_num = df[name].isnull().sum()
    if total_num > 0:
        print ('Number of null of value of {0}: {1}'.format(name,total_num))

df.columns.values
```

Now our training set is consist of 11 cols and 712 rows.
Because our test data is for prediction and output prediction will be submitted on Kaggle, so I will split the training test set to training data and cross-validated data for choose the best algorithm on cross-validated data.
But at first let visualization training set

```python
#plt.grid()

fig = plt.figure(figsize=(20,10), dpi=1600)


alpha=alpha_scatterplot = 0.2 
alpha_bar_chart = 0.55

ax1 = fig.add_subplot(231)
df.Survived.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
ax1.set_xlim(-1, 2)
# puts a title on our graph
ax1.set_title("Distribution of Survival, (1 = Survived)")    
ax1.grid()

ax2 = fig.add_subplot(232)
df.Pclass.value_counts().plot(kind='barh', alpha=alpha_bar_chart)
ax2.set_ylim(-1, len(df.Pclass.value_counts()))
ax2.set_title("Class Distribution")
ax2.grid()

ax3 = fig.add_subplot(233)
df.Sex.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
ax3.set_title('Sex Distribution')
ax3.grid()

ax3 = fig.add_subplot(234)
df.Embarked.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
ax3.set_title('Embarked Distribution')
ax3.grid()

ax4 = fig.add_subplot(235)
ax4.set_xlim(0,90)
df.Age.plot(kind='hist', alpha=alpha_bar_chart)
ax4.grid()
ax4.set_title('Age Distribution')
```

*** Analysis from above visualizations

+ Percentage of survived is less than 50%.
+ Class 3 has largest number of people. (class 3 always at bottom of ship as cheapest ticket)
+ Male is larger than female
+ Embarked at Southampton is the most freqency.
+ People on ship are young as frequency of age 20-30 is largest

### Now let make visualization about collation between features and Surived



```python
fig = plt.figure(figsize=(18,10), dpi=1600)

ax1 = fig.add_subplot(321)
df.Survived[df.Pclass==1].plot(kind='kde')
df.Survived[df.Pclass==2].plot(kind='kde')
df.Survived[df.Pclass==3].plot(kind='kde')
ax1.set_title('Survived Distribution with Pclass')
ax1.set_ylabel('Pclass')
ax1.set_xlabel('Survived')
ax1.legend(['Pclass 1','Pclass 2','Pclass 3'])


ax2 = fig.add_subplot(322)
df_female=df.Survived[df.Sex=='female'].value_counts().sort_index()
df_male = df.Survived[df.Sex=='male'].value_counts().sort_index()

df_male.plot(kind='barh', label='Male',alpha=0.55)
df_female.plot(kind='barh', color='#FA2379',label='Female',alpha=0.55)
ax2.set_title("Who Survived? with respect to Gender, (raw value counts) ");ax2.legend()

ax3 = fig.add_subplot(323)
(df_male/float(df_male.sum())).plot(kind='barh', label='Male',alpha=0.55)
(df_female/float(df_female.sum())).plot(kind='barh', color='#FA2379',label='Female',alpha=0.55)
ax3.set_title("Who Survived proportionally? with respect to Gender"); ax3.legend(loc='best')
```
